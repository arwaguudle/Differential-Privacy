```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Laplace distribution
mean = 0  # Mean
b = 1   # Scale parameter

# Generate 1,000 samples from Laplace distribution
samples = np.random.laplace(mean, b, 1000)

# Plot histogram of the samples
plt.hist(samples, bins=50, density=True, alpha=0.6, color='green')

# Plot Laplace probability density function
x = np.linspace(-10, 10, 1000)
pdf = np.exp(-np.abs(x - mean) / b) / (2.0 * b)
plt.plot(x, pdf, 'r', linewidth=2)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram of Laplace Distribution')
plt.grid(True)
plt.show()
```
